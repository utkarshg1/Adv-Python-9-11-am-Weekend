### List vs Array multiplication of numbers

### Generating normally distributed data

```python
np.random.normal(5.5, 1/3)
```

```python
heights = np.random.normal(loc=5.5, scale=1/3, size=1_00_000)
heights
```

```python
sns.histplot(data=heights, kde=True)
```

```python
np.sin(angle_rad)
```
